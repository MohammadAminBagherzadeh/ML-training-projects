```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
from sklearn import linear_model
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('D:\Programming\AI\Jadi\Practice\Simple_linear_Regression\Salary_dataset.csv')
df.head()
```

```python
X = df.YearsExperience
Y = df.Salary
```

```python
plot.scatter(X, Y)
plot.title('salary to exprience ratio')
plot.xlabel('years of exprience')
plot.ylabel('salary')
plot.show()
```

```python
msk = np.random.rand(len(X)) < 0.8
train = X[msk]
test = X[~msk]
train_Y = Y[msk]
test_Y = Y[~msk]
```

```python
plot.scatter(train, train_Y, color = 'red', alpha = 0.5)
plot.scatter(test, test_Y, color = 'blue', alpha = 0.5)
plot.xlabel('Exprience')
plot.ylabel('Salary')
plot.show()
```

```python
X_train = np.asanyarray(train)
Y_train = np.asanyarray(train_Y)
X_train = X_train.reshape(-1, 1)
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_train)
coef = reg.coef_
intercept = reg.intercept_
print(coef)
print(intercept)
```

```python
plot.scatter(X, Y, color='red', alpha=0.5)
plot.plot(X, coef*(X)+intercept, '-b')
plot.xlabel('Exp')
plot.ylabel('Money')
plot.show()
```

```python
X_test = np.asanyarray(test)
X_test = X_test.reshape(-1,1)
Y_test = np.asanyarray(test_Y)
Y_hat = reg.predict(X_test)
score = r2_score(Y_test, Y_hat)
score_r = np.round(score, 2)
print(score_r)
```
