```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
items_df = pd.read_csv('items.csv')
reviews_df = pd.read_csv('reviews.csv')
```

```python
reviews_df = reviews_df.drop('date', axis=1)
reviews_df = reviews_df.drop('verified', axis=1)
reviews_df = reviews_df.drop('title', axis=1)
reviews_df = reviews_df.drop('body', axis=1)
reviews_df = reviews_df.drop('helpfulVotes', axis=1)
```

```python
reviews_df.head()
```

```python
userInput = [
            {'asin':'B00V7FXCZ2', 'rating':5},
            {'asin':'B00CEKXJ3Y', 'rating':4},
            {'asin':'B07C5X8Q3K', 'rating':5},
            {'asin':'B07JZRL7N5', 'rating':3}
         ]
inputPhone = pd.DataFrame(userInput)
inputPhone
```

```python
userSubset = reviews_df[reviews_df['asin'].isin(inputPhone['asin'].tolist())]
userSubset
```

```python
# topSimilarity=userSubset.sort_values(by='rating', ascending=False)[0:50]
# topSimilarity
userSubsetGroup = userSubset.groupby(['name'])
userSubsetGroup.get_group('Kindle Customer')
```

```python
# mask = (topSimilarity.iloc[:, 2] > (rating-1)) & (topSimilarity.iloc[:, 2] < (rating+1))
# topSimilarity = topSimilarity[~mask]
# topSimilarity
userSubsetGroup = sorted(userSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
userSubsetGroup[0:3]
```

```python
userSubsetGroup = userSubsetGroup[0:100]
```

```python
pearsonCorrelationDict = {}

for name, group in userSubsetGroup:
    group = group.sort_values(by='asin')
    inputPhone = inputPhone.sort_values(by='asin')
    nRatings = len(group)
    temp_df = inputPhone[inputPhone['asin'].isin(group['asin'].tolist())]
    tempRatingList = temp_df['rating'].tolist()
    tempGroupList = group['rating'].tolist()
    Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```

```python
pearsonCorrelationDict.items()
```

```python
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index')
pearsonDF.columns = ['similarityIndex']
pearsonDF['userId'] = pearsonDF.index
pearsonDF.index = range(len(pearsonDF))
pearsonDF.head()
```

```python
topUsers=pearsonDF.sort_values(by='similarityIndex', ascending=False)[0:50]
topUsers['userId'] = topUsers['userId'].str[0].astype(str)
topUsers.head(10)
```

```python
topUsersRating=topUsers.merge(reviews_df, left_on='userId', right_on='name', how='inner')
topUsersRating.head()
```

```python
topUsersRating['weightedRating'] = topUsersRating['similarityIndex']*topUsersRating['rating']
topUsersRating.head()
```

```python
tempTopUsersRating = topUsersRating.groupby('asin').sum()[['similarityIndex','weightedRating']]
tempTopUsersRating.columns = ['sum_similarityIndex','sum_weightedRating']
tempTopUsersRating.head()
```

```python
recommendation_df = pd.DataFrame()
recommendation_df['weighted average recommendation score'] = tempTopUsersRating['sum_weightedRating']/tempTopUsersRating['sum_similarityIndex']
recommendation_df['asin'] = tempTopUsersRating.index
recommendation_df.head()
```

```python
recommendation_df = recommendation_df.sort_values(by='weighted average recommendation score', ascending=False)
recommendation_df.head(10)
```
