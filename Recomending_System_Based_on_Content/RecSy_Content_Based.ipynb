```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('imdb_movies.csv')
df.head()
```

```python
df = df.drop('date_x', axis=1)
df = df.drop('overview', axis=1)
df = df.drop('crew', axis=1)
df = df.drop('orig_title', axis=1)
df = df.drop('orig_lang', axis=1)
df = df.drop('budget_x', axis=1)
df = df.drop('revenue', axis=1)
df = df.drop('country', axis=1)
df = df.drop('status', axis=1)
df = df.drop_duplicates(subset='names')
df.head()
```

```python
df['genre'] = df.genre.str.split(',')
df.head()
```

```python
df = df.dropna(subset=['genre'])
```

```python
movies_df = df.copy()

for index, row in df.iterrows():
    for genres in row['genre']:
        movies_df.at[index, genres] = 1

movies_df = movies_df.fillna(0)
movies_df.head()
```

```python
movies_df = movies_df.drop('genre', axis=1)
movies_df.head()
```

```python
userInput = [
            {'names':'The Super Mario Bros. Movie', 'score':72},
            {'names':'Troll', 'score':65},
            {'names':'Encanto', 'score':78},
            {'names':'Bullet Train', 'score':85},
            {'names':'Sing 2', 'score':68}
         ] 
inputMovies = pd.DataFrame(userInput)
inputMovies
```

```python
userMovies = movies_df[movies_df['names'].isin(inputMovies['names'].tolist())]
userMovies
```

```python
userMovies = userMovies.reset_index(drop=True)
userGenreTable = userMovies.drop('names', axis=1)
userGenreTable
```

```python
userProfile = userGenreTable.transpose().dot(inputMovies['score'])
userProfile
```

```python
genreTable = movies_df.set_index(movies_df['names'])
genreTable = genreTable.drop('names', axis=1).drop('score', axis=1)
genreTable.head()
```

```python
genreTable.shape
```

```python
recommendationTable_df = ((genreTable*userProfile).sum(axis=1))/(userProfile.sum())
recommendationTable_df
```

```python
recommendationTable_df = recommendationTable_df.sort_values(ascending=False)
recommendationTable_df.head()
```
